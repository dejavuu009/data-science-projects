```python
import numpy as np
import pandas as pandas
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = sns.load_dataset('tips')
df.head(10)
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.describe(include='category').T
```

```python
graf = sns.relplot(data=df, x='total_bill', y='tip')
```

```python
graf = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size')
```

```python
graf = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time')
```

```python
graf = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time', row='smoker')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', hue='sex', kind='swarm')
```

```python
_ = sns.catplot(data=df, x='day', hue='sex', kind='count')
```
